### Affichage de la répartition des données selon un couple d'attributs

```python
import matplotlib
from itertools import cycle
import pylab as pl #facilite travail matplotlib
```

```python
def plot_2D(data, target, target_names):
    colors = cycle('cmykwrgb') # cycle de couleurs
    target_ids = range(len(target_names)) #id des classes
    pl.figure()
    for i, c, label in zip(target_ids, colors, target_names): #i:0,1,2
        pl.scatter(data[target == i, 0], data[target == i, 1], c=c, label=label)
    pl.legend()
    pl.plot([2.5,2.5],[0,3],"b--")
    pl.show()
    
plot_2D(irisData.data,irisData.target,irisData.target_names)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="whitegrid")
modeles = {
    "Naive Bayes": lambda: naive_bayes.MultinomialNB(fit_prior=True),
    "KNN": KNeighborsClassifier,
    "Arbre de décision": DecisionTreeClassifier,
    "XGBoost": lambda: XGBClassifier(use_label_encoder=False, verbosity=0, silent=True),
    "Forêt aléatoire": RandomForestClassifier }
resultats = {}

for nom_modele, modele in modeles.items():
    print(f"Modèle: {nom_modele}")

    erreur_map = {}
    
    if nom_modele == "XGBoost":
        tailles_test = np.arange(0.1, 0.9, 0.01)
    else:
        tailles_test = np.arange(0.1, 0.9, 0.05)
    
    for taille_test in tailles_test:
        err = np.array([])
        for _ in range(50):
            donnees_train, donnees_test, cible_train, cible_test = train_test_split(irisData.data, irisData.target, test_size=taille_test)
            m = modele()
            m.fit(donnees_train, cible_train)
            erreur = 1 - m.score(donnees_test, cible_test)
            err = np.append(err, erreur)

        erreur_map[round(taille_test, 2)] = err.mean()

    x = list(erreur_map.keys())
    y = list(erreur_map.values())

    plt.figure(figsize=(8, 4))
    sns.barplot(x=x, y=y, palette="pastel")
    
    if nom_modele == "XGBoost":
        plt.xlabel("Taille de test")
    else:
        plt.xlabel("Pourcentage de test")
    
    plt.ylabel("Erreur")
    plt.title(nom_modele)
    plt.xticks(rotation=90)  
    plt.show()

    taille_test_optimale = min(erreur_map, key=erreur_map.get)
    meilleure_precision = 1 - erreur_map[taille_test_optimale]
    resultats[nom_modele] = {"Taille de test optimale": taille_test_optimale, "Meilleure précision": meilleure_precision}
    print("Taille de test optimale: ", taille_test_optimale)
    print("Meilleure précision: ", meilleure_precision)
    print("\n" + '*' * 50 + "\n")

# Afficher le dictionnaire des résultats sous forme de grille avec une palette de couleurs
resultats_df = pd.DataFrame(resultats).T
plt.figure(figsize=(8, 4))
sns.heatmap(resultats_df, annot=True, cmap="YlGnBu")
plt.title("Résultats des modèles")
plt.show()
```
